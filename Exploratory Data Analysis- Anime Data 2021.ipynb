### Let's look at the Release Seasons!

Anime is a unique and fascinating medium that  differs in it's approach to production and release. Unlike the Western model of full series releases, anime releases are divided into four seasons, each lasting three months, and are categorized according to the weather season: `Winter`, `Spring`, `Summer`, and `Fall`. 

```python
plt.figure(figsize=(7,4), tight_layout=True)
plt.rcParams["axes.edgecolor"] = "black"
plt.rcParams["axes.linewidth"] = 0.50

ax= df.Season.value_counts().sort_values() \
  .plot(kind='barh', color = ['#FFD23F','#33658A','#7BD389','#E56B6F'], title = "In which season are most anime produced?" )
```

<b> Spring </b> is the busiest time for the industry and the audience!

The scatter plot is very crowded, owing to the vast number of datapoints. No apparent correlation can be detected between ratings and year of release. However, we <b> can </b> be sure that throughout the years, mass of anime released has only risen.

```python
df_corr = df[["Episodes", "Rating", "Release_year"]].dropna().corr()
```

```python
ax = sns.heatmap(df_corr, annot= True)
plt.show()
```

There is no correlation present among  any of the features...which makes sense. The number of episode doesnt influence the ratings and vice_versa. Also, the year of release does not have any correlation with others..meaning, both older and newer anime have equal shots at being good.

### Step 5 : Further Questions and Insights 

#### Popularity Indicator 
 Okay, so TV Series and Movies are typical anime types. What about their content quality? Let's judge that by Ratings metric.
 

```python
# Creating categories to measure popularity out of Ratings numeric column
ranges =[0,3,3.5,4.3,5]
labels = ["Watchable","Good", "Great", "Must-watch"]
df["Quality"] = pd.cut(df['Rating'], bins= ranges, labels = labels)
```

```python
sns.set_style('darkgrid') # darkgrid, white grid, dark, white and ticks
plt.rc('axes', titlesize=20)     # fontsize of the axes title
plt.rcParams["axes.edgecolor"] = "black"
plt.rcParams["axes.linewidth"] = 0.50
plt.rc('axes', labelsize=14)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=13)    # fontsize of the tick labels
plt.rc('ytick', labelsize=13)    # fontsize of the tick labels
plt.rc('legend', fontsize=13)    # legend fontsize
plt.rc('font', size=13)


plt.figure(figsize=(10,10))
ax = df.groupby(['Type', 'Quality'])['Name'].count() \
  [["Movie", "TV", "Web"]] \
  .unstack(level=1)\
  .plot(kind='bar', title= "Anime Content by Popularity")
ax.set_ylabel("Number of titles")
ax.set_xlabel("")
plt.show()
```

So, most anime content ratings range from good to great, which attests to it's quality. <br/>
